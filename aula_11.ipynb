```python
#%pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns

gorjetas = sns.load_dataset('tips')

display(gorjetas.head())
print(gorjetas.shape)

print(gorjetas.describe())
```

```python
for i in gorjetas.columns:
    print(i)
```

```python
sns.set_theme(style='whitegrid')
sns.scatterplot(
    data=gorjetas,
    x='total_bill',
    y='tip',
    hue='time',
)
plt.title('relação entre total da conta e gorjeta')
plt.show()
```

```python
# Gráfico de colunas no seaborn

sns.barplot(
    data= gorjetas,
    x='day',
    y='total_bill'
)
plt.title('Pagamento por dias da semana')
plt.show()
```

```python
#Gráfico de histograma

sns.histplot(
    data=gorjetas,
    x='tip',
    kde=True,
)
plt.title('Distribuição das gorjetas')
plt.show()
```

```python
# Dataset - Pinguins

#baixar o dataset -penguins

pinguin = sns.load_dataset('penguins')

display(pinguin.head())

especies = pinguin['species'].unique()
print(especies)

for i in pinguin.columns:
    print(i)
```

```python
sns.scatterplot(
    data=pinguin,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species'
)
plt.show()
```

```python
sns.scatterplot(
    data=pinguin,
    x='body_mass_g',
    y='flipper_length_mm',
    hue='species'
)
plt.show()
```

```python
# Gráfico boxplot

sns.boxplot(
    data=pinguin,
    x='species',
    y='body_mass_g',
    hue='island'
)
plt.title('relação entre as espécies, ilhas e peso')
plt.savefig('grafico_boxplot.png', dpi=300)
plt.show()
```
